```python
from tkinter import *
from tkinter import ttk
from tkinter import filedialog as fd
import os
from lzw.Compress import compress as com
from lzw.Decompress import decompress as dec
import time
import threading


root = Tk()
root.title("Compress and Decompress Text File/ Image")

# Example text
filePath = 'Nhập đường dẫn ở đây ...'


# Define function
def step(opt='com', max_utf_char=11000):
    my_progress.start(1)
    root.update_idletasks()

    time.sleep(1)
    pathToCompress = feet.get()
    parentPath = os.path.abspath(os.path.join(pathToCompress, os.pardir))

    if(opt=='com'):
        task = com(feet.get(),parentPath,encoding='utf-8', max_utf_char=max_utf_char, limit=20000000000)
        task.encode()

        outName, typeName = pathToCompress.split('/')[-1].split('.')[:]
        outFile = os.path.join(os.path.abspath(os.path.join(pathToCompress, os.pardir)), outName + '_compressed.' + typeName)
        out.set(outFile)

        print("Finish")

        outName, typeName = pathToCompress.split('/')[-1].split('.')[:]
        outFile = os.path.join(os.path.abspath(os.path.join(pathToCompress, os.pardir)), outName + '_compressed.' + typeName)
        out.set(outFile)
        infoFileOut.set(file_size(outFile))

        my_progress.stop()
        my_progress['value'] = 0
    else:
        task = dec(feet.get(),parentPath,encoding='utf-8', max_utf_char=max_utf_char)
        task.decode()


        outName, typeName = pathToCompress.split('/')[-1].split('.')[:]
        outFile = os.path.join(os.path.abspath(os.path.join(pathToCompress, os.pardir)), outName + '_decompressed.' + typeName)
        out.set(outFile)

        print("Finish")

        outName, typeName = pathToCompress.split('/')[-1].split('.')[:]
        outFile = os.path.join(os.path.abspath(os.path.join(pathToCompress, os.pardir)), outName + '_decompressed.' + typeName)
        out.set(outFile)
        infoFileOut.set(file_size(outFile))

        my_progress.stop()
        my_progress['value'] = 0


def stop():
    my_progress.stop()
    my_progress['value'] = 0

def convert_bytes(num):
    """
    this function will convert bytes to MB.... GB... etc
    """
    for x in ['bytes', 'KB', 'MB', 'GB', 'TB']:
        if num < 1024.0:
            return "%3.1f %s" % (num, x)
        num /= 1024.0
 
def file_size(file_path):
    """
    this function will return the file size
    """
    if os.path.isfile(file_path):
        file_info = os.stat(file_path)
        return convert_bytes(file_info.st_size)

def select_file():
    filetypes = (
        ('text files', '*.txt'),
        ('text files', '*.docx'),
        ('png images', '*.png'),
        ('jpg images', '*.jpg'),
        ('All files', '*.*')
    )

    filename = fd.askopenfilename(
        title='Open a file',
        initialdir='/',
        filetypes=filetypes)
    
    feet.set(filename)
    infoFile.set(file_size(filename))

    print(infoFile)


def compress():
    threading.Thread(target=step('com')).start()


def decompress():
    threading.Thread(target=step(opt='des')).start()


mainframe = ttk.Frame(root, padding="3 3 12 12")
mainframe.grid(column=0, row=0, sticky=(N, W, E, S))
root.columnconfigure(0, weight=1)
root.rowconfigure(0, weight=1)

feet = StringVar()
feet.set(filePath)
feet_entry = ttk.Entry(mainframe, width=50, textvariable=feet)
feet_entry.grid(column=1, row=1, sticky=(W, E))

out = StringVar()
out_entry = ttk.Entry(mainframe, width=50, textvariable=out)
out_entry.grid(column=3, row=1, sticky=(W, E))


infoFile = StringVar()
ttk.Label(mainframe, textvariable=infoFile).grid(column=1, row=2, sticky=(W,E))

infoFileOut = StringVar()
ttk.Label(mainframe, textvariable=infoFileOut).grid(column=3, row=2, sticky=(W,E))

my_progress = ttk.Progressbar(mainframe, orient=HORIZONTAL,length=100, mode='indeterminate')
my_progress.grid(column=2, row=4)




meters = StringVar()
ttk.Label(mainframe, textvariable=meters).grid(column=2, row=2, sticky=(N, S))

ttk.Button(mainframe, text="Compress", command=compress).grid(column=2, row=2, sticky=W)
ttk.Button(mainframe, text="Decompress", command=decompress).grid(column=2, row=3, sticky=W)

ttk.Button(mainframe, text="Open file", command=select_file).grid(column=2, row=1, sticky=W)
# ttk.Label(mainframe, text="is equivalent to").grid(column=1, row=2, sticky=E)
# ttk.Label(mainframe, text="meters").grid(column=3, row=2, sticky=W)

for child in mainframe.winfo_children(): 
    child.grid_configure(padx=5, pady=5)

feet_entry.focus()

# Short cut
# root.bind("<Return>", calculate)

root.mainloop()
```

```python
def compress_lzw(input_file, output_file):
    # Define initial dictionary with UTF-8 characters
    dictionary = {chr(i): i for i in range(65536)}

    # Read input file
    with open(input_file, 'r', encoding='utf-8') as f:
        input_text = f.read()

    compressed = []
    current_sequence = ""
    for char in input_text:
        # Add current character to current sequence
        new_sequence = current_sequence + char
        if new_sequence in dictionary:
            # If current sequence is in dictionary, add current character to it
            current_sequence = new_sequence
        else:
            # Output the code for the current sequence to the compressed list
            compressed.append(dictionary[current_sequence])
            # Add the new sequence to the dictionary
            dictionary[new_sequence] = len(dictionary)
            # Reset current sequence to current character
            current_sequence = char

        # Check if dictionary size exceeds the maximum limit for integer values
        if len(dictionary) >= 65536:
            dictionary.clear()
            dictionary = {chr(i): i for i in range(65536)}

    # Output the last code to the compressed list
    compressed.append(dictionary[current_sequence])

    # Write the compressed list to output file
    with open(output_file, 'wb') as f:
        for code in compressed:
            f.write(code.to_bytes(2, 'big'))

    print("Compression successful.")

def decompress_lzw(input_file, output_file):
    # Define initial dictionary with UTF-8 characters
    dictionary = {i: chr(i) for i in range(65536)}

    # Read compressed file
    with open(input_file, 'rb') as f:
        compressed = []
        while True:
            # Read two bytes at a time to get code
            bytes_read = f.read(2)
            if not bytes_read:
                break
            code = int.from_bytes(bytes_read, 'big')
            compressed.append(code)

    decompressed = ""
    current_sequence = dictionary[compressed[0]]
    for code in compressed[1:]:
        if code in dictionary:
            # If code is in dictionary, add its value to current sequence
            new_sequence = dictionary[code]
        elif code == len(dictionary):
            # If code is equal to dictionary length, the new sequence is the current sequence + its first character
            new_sequence = current_sequence + current_sequence[0]
        else:
            raise ValueError("Invalid compressed data.")

        # Add the new sequence to the dictionary
        dictionary[len(dictionary)] = current_sequence + new_sequence[0]

        # Check if dictionary size exceeds the maximum limit for integer values
        if len(dictionary) >= 65536:
            dictionary.clear()
            dictionary = {i: chr(i) for i in range(65536)}

        # Add the new sequence to the decompressed string
        decompressed += new_sequence

        # Reset current sequence to new sequence
        current_sequence = new_sequence

    # Write the decompressed string to output file
    with open(output_file, 'w', encoding='utf-8') as f:
        f.write(decompressed)

    print("Decompression successful.")
```

```python
import os
import struct

def compress_lzw_utf8(input_path, output_path):
    # Read input file
    with open(input_path, "r", encoding="utf-8") as input_file:
        data = input_file.read().encode("utf-8")

    # Compress data using LZW
    dictionary = {bytes([i]): i for i in range(256)}
    code = 256
    compressed_data = bytearray()
    current_sequence = b""
    for byte in data:
        sequence = current_sequence + bytes([byte])
        if sequence in dictionary:
            current_sequence = sequence
        else:
            compressed_data += struct.pack(">I", dictionary[current_sequence])
            dictionary[sequence] = code
            code += 1
            current_sequence = bytes([byte])

    if current_sequence:
        compressed_data += struct.pack(">I", dictionary[current_sequence])

    # Write compressed data to output file
    with open(output_path, "wb") as output_file:
        output_file.write(compressed_data)

    # Check if compressed file is smaller than input file
    input_size = os.path.getsize(input_path)
    output_size = os.path.getsize(output_path)
    compression_ratio = output_size / input_size
    if compression_ratio >= 1:
        print("WARNING: Compressed file is not smaller than input file")
    else:
        print("Compression successful: {:.2%} reduction in file size".format(1 - compression_ratio))
```

```python
import struct
from PIL import Image

def decompress_lzw_png(compressed_path, output_path):
    # Read the compressed data from the file
    with open(compressed_path, "rb") as f:
        compressed_data = f.read()

    # Decompress the data using LZW
    MAX_TABLE_SIZE = 2 ** 12
    table = {i: bytes([i]) for i in range(256)}
    next_code = 256
    string = b""
    output_data = b""
    for i in range(0, len(compressed_data), 8):
        code = struct.unpack(">Q", compressed_data[i:i+8])[0]
        if code not in table:
            new_string = string + bytes([string[0]])
        else:
            new_string = table[code]
        output_data += new_string
        if len(table) < MAX_TABLE_SIZE:
            table[next_code] = string + bytes([new_string[0]])
            next_code += 1
        string = new_string

    # Convert the raw image data to a PNG image and save it
    original_image = Image.open("input.png")

    width, height = original_image.size
    print('Width', width, 'Height', height)
    img = Image.frombytes("RGB", (width, height), output_data)
    img.save(output_path, format="PNG")
```

```python
import numpy as np
def decompress_lzw_raw(compressed_file_path, raw_file_path):
    with open(compressed_file_path, "rb") as f:
        compressed_data = f.read()
    
    decompressed_data = decompress_lzw_png(compressed_data)
    with open(raw_file_path, "wb") as f:
        f.write(decompressed_data)

    # Load the raw data as a numpy array
    raw_data = np.frombuffer(decompressed_data, dtype=np.uint8)

    # Reshape the raw data into a 2D array
    img_data = raw_data.reshape((50, 50, 3))

    # Create an image from the data
    img = Image.fromarray(img_data, 'RGB')
    
    return img
```
